```python
a=np.array([1,5,6,8,1,3,8,6])
b=np.array([8,6,1,3,8,1,6,12])
pt.xlabel('a.label') #labeling the x-axis
pt.ylabel('b.label') #labeling the y-axis
pt.plot(a,b,'--g+') #for dotted lines and green colour and + marker
pt.show()
```

```python
a=np.array(['jan','feb','mar','apr','may'])
b=np.array([8,6,1,3,7])
pt.xlabel('months') #labeling the x-axis
pt.ylabel('celsius') #labeling the y-axis
pt.title('temperature')
pt.plot(a,b,'--b*') #for dotted lines and blue colour and + marker
pt.show()
```

```python
a=np.array(['jan','feb','mar','apr','may'])
b=np.array([8,6,1,3,7])
pt.xlabel('months') #labeling the x-axis
pt.ylabel('celsius') #labeling the y-axis
pt.title('temperature')
pt.bar(a,b)
pt.show()
```

```python
a=np.array(['jan','feb','mar','apr','may'])
b=np.array([8,6,1,3,7])
pt.xlabel('celsius') #labeling the x-axis
pt.ylabel('months') #labeling the y-axis
pt.title('temperature')
pt.barh(a,b,color='red',align='edge') #by default align=centre
pt.show()
```
