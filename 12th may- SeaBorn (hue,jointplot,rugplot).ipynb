```python
import pandas as pd
import seaborn as sb
import warnings
warnings.filterwarnings('ignore')
```

```python
tit=pd.read_csv('titanic_train.csv')
tit
```

```python
sb.catplot(x='Pclass',y='Survived',hue='Sex',data=tit,kind='bar',) #hue can only be categpry column
sb.set(style='whitegrid')
```

```python
sb.catplot(x='Pclass',y='Survived',hue='Sex',data=tit,kind='violin',) #hue can only be categpry column
sb.set(style='whitegrid')
```

```python
sb.distplot(tit['Age'],bins=10) #curve is termed kde , if kide=false then curve wont appear
```

```python
sb.kdeplot(tit['Age'])
```

```python
sb.jointplot(x='Age',y='Pclass',data=tit) #jointpl
```

```python
sb.jointplot(x='Age',y='Pclass',data=tit,kind='kde') #kde= kernel density distribution
```

```python
sb.rugplot(x='Age',y='Pclass',data=tit)
```

```python
sb.boxplot(x='Sex',y='Age',hue='Pclass',data=tit,palette='colorblind') #hue can only be categpry column
sb.set(style='whitegrid')
```

```python
sb.pairplot(tit) #shows all the possible graphs using two variables
```

```python
 sb.countplot('Pclass',data=tit) #first variable as treated as xaxis by default / no need to write axis= everytime
```

```python
sb.lmplot(x='Fare',y='Age',hue='Embarked',data=tit,palette='colorblind') #hue can only be categpry column
sb.set(style='whitegrid')
```
